## PRECIPITATION DATA

```python
import pandas as pd
import glob

# Create an empty list to store DataFrames
precip_list = []

# Read each CSV file and append to the list
for filename in glob.glob('C:/Users/interactwel/Desktop/PRISM datasets/excel_to_text_script_folder/precipitation_data/*.csv'):
    df = pd.read_csv(filename, index_col='date_prism', parse_dates=['date_prism'])
    precip_list.append(df)

# Concatenate the DataFrames along the columns axis
precip = pd.concat(precip_list, axis='columns')

# Display the first 100 rows
print(precip.head(100))
```

```python

# List of file names
file_names = [r"C:\Users\interactwel\Desktop\PRISM datasets\excel_to_text_script_folder\precipitation_data\precipitation_2018.csv",
               r"C:\Users\interactwel\Desktop\PRISM datasets\excel_to_text_script_folder\precipitation_data\precipitation_2019.csv", 
               r"C:\Users\interactwel\Desktop\PRISM datasets\excel_to_text_script_folder\precipitation_data\precipitation_2020.csv",
                 r"C:\Users\interactwel\Desktop\PRISM datasets\excel_to_text_script_folder\precipitation_data\precipitation_2021.csv", 
                 r"C:\Users\interactwel\Desktop\PRISM datasets\excel_to_text_script_folder\precipitation_data\precipitation_2022.csv"
                 ]

# Read the first CSV file
df = pd.read_csv(file_names[0])


```

```python
# Loop through the rest of the files and concatenate them
for file in file_names[1:]:
    temp_df = pd.read_csv(file)
    df = pd.concat([df, temp_df], ignore_index=True)
```

```python
# Sort the DataFrame by date if necessary
df = df.sort_values(by='date_prism')
```

```python
df.head(100)
```

```python
# Save the merged DataFrame to a new CSV file
df.to_csv("Precipitation_data_big_watershed.csv", index=False)
```
